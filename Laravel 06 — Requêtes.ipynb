# La classe Request

## Introduction

La classe `Illuminate\Http\Request` de Laravel offre un moyen orienté objet d'interagir avec la requête HTTP en cours de traitement par votre application et de récupérer les données, les cookies et les fichiers qui ont été soumis avec la requête.

L'instance de cette classes est construite au tout début du cycle par le routeur.

La classe `Request` de Laravel est une extension de la classe `Request` de Symfony.

## Interagir avec la requête

### Appeler l'objet

L'objet `Request` étant juste un objet, son rôel est principalemnt d'être appelé dans des services qui ont beson d'accéder aus données et méta-données d la requête.

Typiquement, les contrôleurs ont souvent besoin deces informations.

Pour cela, nous allons recourir à l'injection de dépendance :

```python
%%php

public function store(Request $request)
{
    $name = $request->input('name');
}
```

Dans cet exemple la métode `store`d'une classe de contrôleurs récupère la requête par injection de dpéendances et accède ensuite à ses propriétés (ici `input` suggère qu'unformulaire a été envoyé.

### Utilitaires

La classe `Request`possède de nombreuses méthodes permettant d'analyser la requête.

#### url

Récupère l'URL de la requête

#### path

Récupère le chemin de l'URL (c'est-à-dire dans le domaine)

#### is

Vérifie si le chemin correspond à une expression régulière donnée

#### routeIs

Vérifie si le nom de la route correspond à une expression régulière donnée

#### méthod

Récupère le nom de la méthode HTTP de la requête

#### ip

Récupère l'adresse IP de l'utilsateur

### Entêtes HTTP

#### header

Rend la vleur d'un entête particulier

#### hasHeader

Vérifie la présence d'un entête dans la requête

#### cookies

Récupères les cookies envoyés par le client

### Négociation de contenu

#### accepts

Rend le contenu de l'entête `Accept`

#### prefers

Rend le type MIME préféré envoyé dans l'entête `Accept`

#### expectsJson

Vérifie que le client attend une réponse au format JSON

### Le corps de la requête

#### all

Récupère toutes les données

#### collect

Idem, mais rend un objet de la classe `Illuminate\Support\Collection`

#### input

`input` est une méthode qui permet de trouver des données envoyées dans le corps de la requête, c'est-à-dire, souvent une reuqête de type `POST`.

La méthode `input` admet le passage d'une valeur par défaut.

#### La chaîne de paramètres

Dans le cas de requêtes de type `GET`, les données sont contenues dans la chaîne de paramètres de l'URL. Onpeut y avoir accès avec la méthode `query`.

### Fichiers

Les ichiers envoyés avec lesrequêtes, typiquement par le biais de formulaires, sont traités séparément.

#### file

Récupère, sous la forme d'un objet, les données d'un fichier téléversé avec la requête.

L'objet résultant possède lui-même des méthodes comme `isValid()`, `path()` qui donnent accès aux détails du fichier ou `store()`qui enregistre le fichier dans le système de fichiers du serveur.

### Utilisation de la session

Il est possible de conserver les information de la requête (les données) dans la session, ce qui peut être très utile encas d'erreur ou autre. L'objet `Request` possède dex méthodes.

#### flash

En registre les données dans la session

#### old

Extrait des données de la session
